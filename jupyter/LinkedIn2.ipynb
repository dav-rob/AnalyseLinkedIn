```python
import sqlite_utils
```

```python
import pandas as pd
```

```python
db = sqlite_utils.Database("../linkedIn_App/linkedIn_Analysis.db")
```

```python
query = """
SELECT 
              job_id,
              job_id, job_title, company, job_details,
              json_extract(analysis, '$."Summary".match_score') AS match_score,
              json_extract(analysis, '$."Summary".strengths') AS strengths,
              json_extract(analysis, '$."Summary".gaps') AS gaps,
              json_extract(analysis, '$."Summary".detailed_analysis') AS detailed_analysis,
               footer_info
          FROM jobs
          ORDER BY match_score DESC
"""
rows = db.query(query)
df=pd.DataFrame(list(rows))
df.head(25)
```

```python
query = """
SELECT 
              *
          FROM jobs
"""
rows = db.query(query)
df=pd.DataFrame(list(rows))
df.head(5)
```
